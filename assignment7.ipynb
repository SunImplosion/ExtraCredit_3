This code performs data-cleaning of HR dataset.

## Instructions

1. **Read the data:** read the data from "HR_5000 Records.csv" using pandas library.

2. **Create a subset-data frame** Create a new data frame that only includes "Emp ID, first name, last name, gender and email". 

3. **Data Engineering:** Use Lambda function to create a new column called "domain". 

4. **Data Visualization:** Use bar-chart to display the distribution of the domain in the given HR dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_raw = pd.read_csv('HR_5000 Records.csv')                         #reads

df_info = df_raw[["Emp ID","First Name","Last Name","Gender","E Mail"]].copy()  #Part 2
df_email = df_raw['E Mail']                                         #creates 'list' of only emails from the file


email = list(df_email)                                              #turns it into list format

domain = (lambda a: str(a[a.find("@")+1: a.find(".",(a.find('.')+1))]))  #creates function given persons email returns domain

d = [*map(domain, email)]                                           #For all elements in email get domain

xax = []
for i in d :                                                        #gets labels for the x-axis
    ex = False
    for a in xax : 
        if a == i:
            ex = True
    if ex == False:
        xax.append(i)
yax = []
for t in xax:                                                       #gets label for the y-axis
    dem = 0
    for it in d:
        if it == t:
            dem+=1
    yax.append(dem)
df_info["Domain"] = d
#plt.figure(figsize=(14,6))                                          #sizes graph
#plt.bar(xax, yax,1)                                                 #greates bar graph
#plt.xticks(rotation=45)                                             #rotates x-ticks
#plt.show()                                                          #shows graph



df_info
```
